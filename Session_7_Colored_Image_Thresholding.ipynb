# Mandatory Task - 1 Take an image find LB and UB and make the binary mask for that colored object

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

### Converting into HSV

```python
image_path = "/content/drive/MyDrive/Colab Notebooks/Balloon-Flower-Growing-in-the-Garden.jpg"

image =  cv2.imread(image_path)

#converting into HSV and RGB

rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

plt.figure(figsize=(10,6))
plt.subplot(1, 2, 1)
plt.imshow(rgb_image)

plt.subplot(1, 2, 2)
plt.imshow(hsv_image)
```

### Creating Binary Mask

```python
lower_yellow = np.array([125, 50, 50]) # lower bound for yellow color that we want to separte as mask
upper_yellow = np.array([160, 255, 255]) # upper bound for yellow color

Binary_mask = cv2.inRange(hsv_image, lower_yellow, upper_yellow)

plt.figure(figsize=(10,6))
plt.subplot(1, 2, 1)
plt.imshow(rgb_image)
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(Binary_mask, cmap='gray')
plt.title("Binary Mask")
```

# Optional Task - 1 Apply the binary mask for MT-1

```python
Masked_image = cv2.bitwise_and(rgb_image, rgb_image, mask=Binary_mask)

plt.figure(figsize=(10,6))

plt.subplot(1, 3, 1)
plt.imshow(rgb_image)
plt.title("Original Image")

plt.subplot(1, 3, 2)
plt.imshow(Binary_mask, cmap='gray')
plt.title("Binary Mask")

plt.subplot(1, 3, 3)
plt.imshow(Masked_image)
plt.title("Masked Image")

```

# Mandatory Task - 2 Take an image, define roi and make a binary mask for that image

```python
image_path2 = "/content/drive/MyDrive/Colab Notebooks/camera.jpg"
image2 = cv2.imread(image_path2)

rgb_image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

# cropping the image for a good shaped image

y1, y2, x1, x2 = 100, 600, 200, 850

cropped_image2 = rgb_image2[y1:y2, x1:x2]

plt.figure(figsize=(10,6))
plt.subplot(1, 2, 1)
plt.imshow(rgb_image2)


plt.subplot(1, 2, 2)
plt.imshow(cropped_image2)
```

### Defining Region of Interest (ROI) and creating mask

```python
mask = np.zeros(cropped_image2.shape[:2], dtype=np.uint8)

mask_x1, mask_x2, mask_y1, mask_y2 = 100, 570, 170, 380

cv2.rectangle(mask, (mask_x1, mask_y1), (mask_x2, mask_y2), 255, -1)

plt.imshow(mask, "gray")
```

# Optional Task - 2 Apply the binary mask for MT-2

```python
#applying mask on the rgb image

masked_image2 = cv2.bitwise_and(cropped_image2, cropped_image2, mask=mask)
plt.imshow(masked_image2)
```

```python
plt.figure(figsize=(10,6))

plt.subplot(1, 3, 1)
plt.imshow(cropped_image2)
plt.title("Original Image")

plt.subplot(1, 3, 2)
plt.imshow(mask, cmap='gray')
plt.title("Binary Mask")

plt.subplot(1, 3, 3)
plt.imshow(masked_image2)
plt.title("Masked Image")
```
