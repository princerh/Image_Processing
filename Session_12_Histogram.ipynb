# **MT: Apply any histogram equalization technique to enhance the features of the attached x-ray image.**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = cv2.imread("/content/drive/MyDrive/Colab Notebooks/x-ray-1.png")
image_rgb = cv2.cvtColor(image_path, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

```python
# histogram of this original image

plt.hist(image_rgb.ravel(), bins=256, range=(0,255), color="k")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency of pixels")
plt.title("Histogram of Original Image")
```

```python
# plotting the histograms of each color channels

plt.figure(figsize=(12, 4))
plt.subplot(1,3,1)
plt.hist(image_rgb[:, :, 0].ravel(), bins=256, range=(0,256), color="r")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.title("Red channel Histogram")

plt.subplot(1, 3, 2)
plt.hist(image_rgb[:, :, 1].ravel(), bins=256, range=(0, 256), color='g')
plt.title('Green Channel Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

plt.subplot(1, 3, 3)
plt.hist(image_rgb[:, :, 2].ravel(), bins=256, range=(0, 256), color='b')
plt.title('Blue Channel Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
```

## **CLAHE - Contrast Limited Adaptive Histogram Equalization**

```python
img_gray = cv2.cvtColor(image_path, cv2.COLOR_BGR2GRAY) # taking gray image , as it is convenient to compare.
```

```python
clahe = cv2.createCLAHE(clipLimit = 10.0, tileGridSize=(8,8))
clahe_img = clahe.apply(img_gray)

plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.imshow(img_gray, cmap='gray')
plt.title('Original Image')


plt.subplot(2, 2, 2)
plt.imshow(clahe_img, cmap='gray')
plt.title('CLAHE')

```

```python
plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.hist(img_gray.ravel(), bins=256, range=(0,255), color="k")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.title("Histogram of original image")

plt.subplot(1,2,2)
plt.hist(clahe_img.ravel(), bins=256, range=(0,255), color="k")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.title("Histogram of CLAHE image")
```

# **OT: Apply different histogram equalization techniques on color image.**

```python
# Applying Global Histogram Equalization (GHE)
ghe_image = cv2.equalizeHist(img_gray)

# Applying adaptive histogram eqalization
ahe = cv2.createCLAHE(clipLimit=1.0, tileGridSize=(8, 8))
ahe_image = ahe.apply(img_gray)

# Applying Contrast Limited Adaptive Histogram equalization
clahe = cv2.createCLAHE(clipLimit=10.0, tileGridSize=(8, 8))
clahe_img = clahe.apply(img_gray)

plt.figure(figsize=(18, 10))

plt.subplot(1, 4, 1)
plt.imshow(img_gray, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 4, 2)
plt.imshow(ghe_image, cmap='gray')
plt.title('Global Histogram Equalization (GHE)')

plt.subplot(1, 4, 3)
plt.imshow(ahe_image, cmap='gray')
plt.title('AHE')

plt.subplot(1, 4, 4)
plt.imshow(clahe_img, cmap='gray')
plt.title('CLAHE')
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.hist(img_rgb[:, :, 0].ravel(), bins=256, range=(0, 256), color='r')
plt.title('Red Channel Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

plt.subplot(1, 3, 2)
plt.hist(img_rgb[:, :, 1].ravel(), bins=256, range=(0, 256), color='g')
plt.title('Green Channel Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

plt.subplot(1, 3, 3)
plt.hist(img_rgb[:, :, 2].ravel(), bins=256, range=(0, 256), color='b')
plt.title('Blue Channel Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply histogram equalization
equalized_image = cv2.equalizeHist(img_gray)


plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(img_gray, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Equalized Image')
plt.imshow(equalized_image, cmap='gray')
plt.axis('off')

plt.show()
```

```python
# histogram comparison of original and histogram equalized image
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.hist(img_gray.ravel(), bins=256, range=(0, 256), color='k')
plt.title('Original Image')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(equalized_image.ravel(), bins=256, range=(0, 256), color='k')
plt.title('Histogram Equalized Image')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
```
