# **Task-1**

```python
import cv2
from matplotlib import pyplot as plt
```

```python
#image loading
image_path = "/content/drive/MyDrive/Colab Notebooks/shouts-animals-watch-baby-hemingway.jpg"
image = cv2.imread(image_path)

#converting image from BGR to RGB (original)
original_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

vertically_flipped = cv2.flip(original_image, 2)
horizontally_flipped = cv2.flip(original_image, 0)
both_flipped = cv2.flip(original_image, -4)

plt.figure(figsize=(12,8))

plt.subplot(2, 2, 1)
plt.imshow(original_image)
plt.title("Original Image")

plt.subplot(2, 2, 2)
plt.imshow(vertically_flipped)
plt.title("Vertically Flipped")

plt.subplot(2, 2, 3)
plt.imshow(horizontally_flipped)
plt.title("Horizontally Flipped")

plt.subplot(2, 2, 4)
plt.imshow(both_flipped)
plt.title("Vertically and Horizontally Flipped")

plt.show()
```

# **Task-2**

```python
#Rotation Method 1
rotated_90_clockwise = cv2.rotate(original_image, cv2.ROTATE_90_CLOCKWISE)
rotated_180_clockwise = cv2.rotate(original_image, cv2.ROTATE_180)
rotated_270_clockwise = cv2.rotate(original_image, cv2.ROTATE_90_COUNTERCLOCKWISE)

#Rotation Method 2

#determining center
(h, w) = original_image.shape[:2]
center = (h // 2, w // 2)

#making a rotation matrix
R1 = cv2.getRotationMatrix2D(center, -45, 1.0)
R2 = cv2.getRotationMatrix2D(center, -115, 1.0)
R3 = cv2.getRotationMatrix2D(center, -225, 1.0)

#applying rotation to orginal image
rotated_45_counterClockwise = cv2.warpAffine(original_image, R1, (w,h))
rotated_115_counterClockwise = cv2.warpAffine(original_image, R2, (w,h))
rotated_225_counterClockwise = cv2.warpAffine(original_image, R3, (w,h))


plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.imshow(rotated_90_clockwise)
plt.title("Rotated 90 degree clockwise")

plt.subplot(2, 3, 2)
plt.imshow(rotated_180_clockwise)
plt.title("Rotated 180 degree clockwise")

plt.subplot(2, 3, 3)
plt.imshow(rotated_270_clockwise)
plt.title("Rotated 270 degree clockwise")

plt.subplot(2, 3, 4)
plt.imshow(rotated_45_counterClockwise)
plt.title("Rotated 45 counter clockwise")

plt.subplot(2, 3, 5)
plt.imshow(rotated_115_counterClockwise)
plt.title("Rotated 115 counter clockwise")

plt.subplot(2, 3, 6)
plt.imshow(rotated_225_counterClockwise)
plt.title("Rotated 225 counter clockwise")

plt.show()
```

# **Task-3**

```python
rescalling = cv2.resize(original_image, None, fx = .5, fy = .5, interpolation = cv2.INTER_CUBIC)
resizing = cv2.resize(original_image, (100,100))

plt.figure(figsize=(12,8))

plt.subplot(1, 3, 1)
plt.imshow(original_image)
plt.title("Original Image")

plt.subplot(1, 3, 2)
plt.imshow(rescalling)
plt.title("Rescalling Image")

plt.subplot(1, 3, 3)
plt.imshow(resizing)
plt.title("Resizing Image")

plt.show()
```

# **Non Mandatory Task**

```python
R, G, B = cv2.split(original_image)

plt.figure(figsize=(12,8))

plt.subplot(1, 4, 1)
plt.imshow(original_image)
plt.title("Original RGB Image")

plt.subplot(1, 4, 2)
plt.imshow(R, cmap="Reds")
plt.title("Red Channel")

plt.subplot(1, 4, 3)
plt.imshow(G, cmap="Greens")
plt.title("Green Channel")

plt.subplot(1, 4, 4)
plt.imshow(B, cmap="Blues")
plt.title("Blue Channel")

plt.show()
```

```python
B, G, R = cv2.split(image)

plt.figure(figsize=(12,8))

plt.subplot(1, 4, 1)
plt.imshow(image)
plt.title("Original BGR Image")

plt.subplot(1, 4, 2)
plt.imshow(B, cmap="Blues")
plt.title("Blue Channel")

plt.subplot(1, 4, 3)
plt.imshow(G, cmap="Greens")
plt.title("Green Channel")

plt.subplot(1, 4, 4)
plt.imshow(R, cmap="Reds")
plt.title("Red Channel")

plt.show()
```

```python
# Convert the BGR image to HSV
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

hue_channel, saturation_channel, value_channel = cv2.split(hsv_image)

plt.figure(figsize=(12, 8))

plt.subplot(1, 4, 1)
plt.imshow(hsv_image)
plt.title("Original HSV Image")

plt.subplot(1, 4, 2)
plt.imshow(hue_channel, cmap='hsv')
plt.title("Hue Channel")

plt.subplot(1, 4, 3)
plt.imshow(saturation_channel, cmap='gray')
plt.title("Saturation Channel")

plt.subplot(1, 4, 4)
plt.imshow(value_channel, cmap='gray')
plt.title("Value Channel")

plt.show()
```
