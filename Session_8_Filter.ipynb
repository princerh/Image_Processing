# Mandatory Task - 1 Contribution of filtering in the given applications

Contribution of Filtering in Various Applications:



1.   Noise Reduction
*  Filtering helps remove unwanted noise from images while preserving essential details.
*  Techniques like Gaussian, Median, and Bilateral filtering smoothen images by reducing high-frequency noise.
*  Essential for enhancing image quality in medical imaging, satellite imagery, and digital photography.



2.  Edge Detection
* Filters like Sobel, Prewitt, and Canny highlight edges by detecting intensity changes.
* Crucial for identifying object boundaries in computer vision applications.
* Used in facial recognition, automated inspection, and robotics for precise contour detection.

3. Feature Extraction
* Filtering enhances specific image features such as textures, corners, and patterns.
* Techniques like Gabor filters, wavelets, and Fourier transforms are used for feature enhancement.
* Plays a key role in biometric authentication, pattern recognition, and AI-based classification tasks.

4. Image Segmentation
* Filters assist in segmenting regions by enhancing contrast and suppressing irrelevant details.
* Used in thresholding-based, region-based, and clustering-based segmentation methods.
* Vital for medical imaging (tumor detection), remote sensing (land classification), and autonomous driving (lane detection).

5. Object Recognition
* Filtering helps preprocess images to improve object recognition accuracy.
* Techniques like template matching, convolutional filters, and deep learning feature maps enhance object visibility.
* Used in AI applications such as facial recognition, self-driving cars, and industrial automation.



# Mandatory Task - 2 Make a custom filter and apply it on an image. Explain it's purpose.

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
image2_path = "/content/drive/MyDrive/Colab Notebooks/blackguy.jpg"
img2 = cv2.imread(image2_path)
image2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
# Edge detection filter defining

array = [
     [1,  0, -1],
    [0,  0,  0],
    [-1, 0,  1]
]

detection_kernel = np.array(array, dtype= np.float32)
print(detection_kernel)
```

```python
detected_image = cv2.filter2D(image2, -1, detection_kernel)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image2)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(detected_image)
plt.title("Edge Detected Image")
plt.axis("off")

plt.show()
```

### Sharpening filter purpose

* It enhances edges along a diagonal direction, unlike simple horizontal/vertical edge filters.
* It does not blur the image, making it useful for feature extraction.
* It works well for object boundary detection, which is crucial in computer vision tasks.

# Optional Task - 1 Apply Gaussian blur on an image

```python
image_path = "/content/drive/MyDrive/Colab Notebooks/apple.jpg"

img = cv2.imread(image_path)
image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


blur = cv2.GaussianBlur(image, (9, 9), 0)

# Display the images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(blur)
plt.title("Gaussian Blur")
```
