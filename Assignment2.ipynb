**Task-1** How image thresholding helps in each application mentioned in the materials?

Image thresholding simplifies image data by converting grayscale or colored images into binary images. This is extremely useful in applications that require distinguishing specific regions, objects, or features from the background. Some contributions of image thresholding are as follows:

1. Object Detection:
Thresholding isolates objects from the background by creating a binary image, where objects of interest (e.g., those above the threshold intensity) are marked as white, and the background is marked as black.
This simplification makes it easier to detect and localize objects in cluttered images.
Example:
Detecting objects in industrial quality control (e.g., identifying defective parts).

2. Segmentation:
Thresholding is often the first step in segmentation, dividing an image into regions of interest. It can differentiate between different materials, tissues, or objects based on intensity levels.
Example: Medical imaging (e.g., segmenting tumors or organs in MRI or CT scans).

3. Edge Detection:
Thresholding is applied after edge detection algorithms (like Sobel or Canny) to refine the edge map, keeping only the most prominent edges while discarding weaker ones. This reduces noise and highlights significant edges.
Example: Detecting boundaries in satellite images or contour extraction for 3D modeling.

4. Image Enhancement:
Enhances image contrast by separating bright and dark regions, making specific areas of interest more visually prominent. Thresholding can also help reduce noise by suppressing unwanted background information.
Example: Enhancing the visibility of text in scanned documents.

5. Pattern Recognition:
Simplifies complex images into binary formats, making it easier for pattern recognition algorithms to identify shapes, textures, or specific features. In machine learning, thresholded images serve as preprocessed input for feature extraction.
Example: Recognizing handwritten digits or characters in Optical Character Recognition (OCR) systems.

```python
import cv2
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('/content/drive/MyDrive/Colab Notebooks/1688919468874.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap="gray")
```

```python
ret, thresholded1 = cv2.threshold(image, 100, 255, cv2.THRESH_BINARY) # if pixel >= threshold, set 255, otherwise set 0
plt.imshow(thresholded1 , cmap="gray")
```

```python
ret, thresholded2 = cv2.threshold(image, 100, 255, cv2.THRESH_BINARY_INV) # if pixel >= threshold, set 0, otherwise set 255
plt.imshow(thresholded2 , cmap="gray")
```

```python
image2 = cv2.imread('/content/drive/MyDrive/Colab Notebooks/shouts-animals-watch-baby-hemingway.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(image2, cmap="gray")
```

```python
ret, thresholded3 = cv2.threshold(image2, 150, 255, cv2.THRESH_TRUNC) # if pixels > threshold then set threshold, otherwise set pixels
plt.imshow(thresholded3, cmap="gray")
plt.show()
print(ret)
```

```python
image3 = cv2.imread('/content/drive/MyDrive/Colab Notebooks/therapy-dog-body-language-meaning-scaled.jpg', cv2.IMREAD_GRAYSCALE)
croppedimage3 = image3[ : , 1000:2500]
plt.imshow(croppedimage3, cmap="gray")
```

```python
ret, thresholded4 = cv2.threshold(croppedimage3, 140, 255, cv2.THRESH_TOZERO)
plt.imshow(thresholded4, cmap="gray")
plt.show()
print(ret)
```

```python
ret, thresholded5 = cv2.threshold(croppedimage3, 140, 255, cv2.THRESH_TOZERO_INV)
plt.imshow(thresholded5, cmap="gray")
plt.show()
print(ret)
```
